```python
data_jan=pd.read_excel('Peminjaman buku 2016.xls',sheetname='Jan',header=3)
data_jan=data_jan[3:]
data_jan=data_jan.reset_index(drop=True)
data_jan
```

```python
data_jan=data_jan[['ID Anggota','ID Buku','Judul']]
data_jan.head()
```

## Hilangkan NAN dan duplicat  

```python
#hilangkan nan
#data_jan=data_jan.dropna(how='any').head(10)
data_jan=data_jan.dropna(how='any')
data_jan.head(10)
```

```python
#https://chrisalbon.com/python/data_wrangling/pandas_selecting_rows_on_conditions/
#nu=ci.loc[~ci['ID Anggota'].duplicated()]
#nu.drop_duplicates()
#HILANGKAN DUPLIKAT
#ci.loc[~ci.duplicated()==True]
data_jan=data_jan.loc[~data_jan.duplicated()==True]
data_jan.head(10)
```

```python
data_iseng=data_jan.copy()
data_iseng=data_iseng[['ID Anggota','ID Buku','Judul','Aktifitas']]
data_iseng=data_iseng.dropna(how='any')
event_type = {
   'Borrow': 1.0,
   'Return': 1.0, 
   'Renew': 2.0  
}

data_iseng['score'] = data_iseng['Aktifitas'].apply(lambda x: event_type[x])
data_iseng.head()
```

```python
item_popularity_df = data_iseng.groupby('Judul')['score'].sum().sort_values(ascending=False).reset_index()
item_popularity_df.head(10)
```
